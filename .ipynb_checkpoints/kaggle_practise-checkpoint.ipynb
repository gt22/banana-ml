```python
def season(x):
    if x in [12, 1, 2]:
        return 'winter'
    elif x in [3, 4, 5]:        
        return 'spring'
    elif x in [6, 7, 8]:        
        return 'summer'
    else:
        return 'fall'

df['Season'] = df.Month.map(season)
```

```python
df.head()
```

```python
pd.get_dummies(df.loc[:, ['Season', 'UniqueCarrier']], columns=['Season','UniqueCarrier'], dummy_na=False).head()
```

```python
df['Velocity'] = df.Distance/(df.ArrTime - df.DepTime)
```
